```python
#Rosalind 11

# Import the files module from google.colab
from google.colab import files

# Step 1: Upload your Rosalind dataset
# This will open a file upload dialog in Colab
uploaded = files.upload()

# Step 2: Read the uploaded file
# uploaded is a dictionary: keys are filenames, values are file contents
# We'll get the first uploaded file
filename = list(uploaded.keys())[0]

# Read DNA string and clean it
with open(filename, 'r') as file:
    dna_text = "".join(file.read().split()).upper()  # remove whitespace/newlines and capitalize

# Set k
k = 14  # replace with your Rosalind problem's k value

# Function to find all most frequent k-mers in a DNA string
def most_frequent_kmers(text, k):
    """
    Finds all k-length substrings (k-mers) in the DNA string `text`
    that occur the most frequently.

    Parameters:
    text (str): DNA sequence (string of A, C, G, T)
    k (int): length of k-mer

    Returns:
    list: list of most frequent k-mers (order does not matter)
    """
    # Dictionary to store counts of each k-mer
    kmer_counts = {}

    # Loop over every possible starting position of a k-mer
    for i in range(len(text) - k + 1):
        current_kmer = text[i:i+k]  # Extract k-mer of length k
        if current_kmer in kmer_counts:
            kmer_counts[current_kmer] += 1  # Increment count if already exists
        else:
            kmer_counts[current_kmer] = 1   # Initialize count if new k-mer

    # Find the maximum frequency among all k-mers
    max_count = max(kmer_counts.values())

    # Collect all k-mers that have the maximum frequency
    most_frequent = []
    for kmer, count in kmer_counts.items():
        if count == max_count:
            most_frequent.append(kmer)

    return most_frequent
result = most_frequent_kmers(dna_text, k)
print("Most frequent", k, "-mers:", result)
```

```python
#Rosalind 12

# Import the files module for uploading files in Colab
from google.colab import files

# Step 1: Upload your Rosalind dataset
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Take the first uploaded file

# Step 2: Read the Genome string from the file
with open(filename, 'r') as file:
    # Remove any whitespace or newlines, and convert to uppercase
    Genome = "".join(file.read().split()).upper()

# Step 3: Set your parameters (replace with values from your Rosalind problem)
k = 9  # length of k-mer
L = 591 # window length
t = 18  # minimum number of occurrences in window

# Step 4: Define the function to find (L, t)-clumps
def find_clumps(Genome, k, L, t):
    """
    Finds all distinct k-mers forming (L, t)-clumps in the Genome.

    Parameters:
    Genome (str): DNA sequence
    k (int): length of k-mer
    L (int): window size
    t (int): minimum number of times k-mer must appear in a window

    Returns:
    set: set of k-mers forming (L, t)-clumps
    """
    clumps = set()  # To store k-mers that form clumps

    # Slide a window of length L along the Genome
    for i in range(len(Genome) - L + 1):
        window = Genome[i:i+L]  # Extract current window
        window_counts = {}       # Dictionary to count k-mers in this window

        # Count k-mers in the current window
        for j in range(L - k + 1):
            kmer = window[j:j+k]
            window_counts[kmer] = window_counts.get(kmer, 0) + 1

        # Add k-mers meeting the threshold t to the clumps set
        for kmer, count in window_counts.items():
            if count >= t:
                clumps.add(kmer)

    return clumps

# Step 5: Run the function and print the results
result = find_clumps(Genome, k, L, t)
print("Number of clump-forming k-mers:", len(result))
print("Clump-forming k-mers:", result)
```

```python
#Rosalind 13

# Import the files module for uploading files in Colab
from google.colab import files

# Step 1: Upload your Rosalind dataset
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Take the first uploaded file

# Step 2: Read the Genome string from the file
with open(filename, 'r') as file:
    Genome = "".join(file.read().split()).upper()  # remove whitespace/newlines and capitalize

# Step 3: Define the function to find minimum skew positions
def minimum_skew_positions(Genome):
    """
    Finds all positions i minimizing Skew(Prefix_i(Genome)).

    Skew(Prefix_i) = #G - #C in the first i nucleotides.

    Parameters:
    Genome (str): DNA sequence

    Returns:
    list: positions i (0-indexed) where skew is minimal
    """
    skew = 0
    skew_list = [0]  # Skew of prefix length 0

    # Compute skew for each prefix
    for nucleotide in Genome:
        if nucleotide == 'G':
            skew += 1
        elif nucleotide == 'C':
            skew -= 1
        # A and T do not change skew
        skew_list.append(skew)

    # Find the minimum skew value
    min_skew = min(skew_list)

    # Collect all positions with minimum skew
    min_positions = [i for i, value in enumerate(skew_list) if value == min_skew]

    return min_positions

# Step 4: Run the function and print results
result = minimum_skew_positions(Genome)
print("Positions minimizing skew:", result)
```

```python
#Rosalind 15

# Import the files module for uploading files in Colab
from google.colab import files

# Step 1: Upload your Rosalind dataset
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Take the first uploaded file

# Step 2: Read Pattern and d from the file
with open(filename, 'r') as file:
    lines = file.read().split()
    Pattern = lines[0].strip().upper()  # DNA string
    d = int(lines[1].strip())           # Maximum mismatches

# Step 3: Define a function to compute Hamming distance
def hamming_distance(s1, s2):
    """
    Computes the Hamming distance between two strings of equal length.
    """
    return sum(1 for a, b in zip(s1, s2) if a != b)

# Step 4: Define the recursive function for Neighbors
def Neighbors(Pattern, d):
    """
    Generates all d-neighbors of a DNA string Pattern.

    Parameters:
    Pattern (str): DNA string
    d (int): maximum number of mismatches

    Returns:
    set: all strings differing from Pattern in at most d positions
    """
    if d == 0:
        return {Pattern}
    if len(Pattern) == 1:
        return {'A', 'C', 'G', 'T'}

    Neighborhood = set()
    Suffix = Pattern[1:]
    SuffixNeighbors = Neighbors(Suffix, d)

    for text in SuffixNeighbors:
        if hamming_distance(Suffix, text) < d:
            for nucleotide in ['A', 'C', 'G', 'T']:
                Neighborhood.add(nucleotide + text)
        else:
            Neighborhood.add(Pattern[0] + text)

    return Neighborhood

# Step 5: Run the function and print the results
neighbors = Neighbors(Pattern, d)
print("Number of neighbors:", len(neighbors))
# Sort the neighbors alphabetically
sorted_neighbors = sorted(neighbors)

# Print each neighbor on a new line (column format)
for neighbor in sorted_neighbors:
    print(neighbor)
```

```python
#Rosalind 16

# Step 1: Import the file upload tool
from google.colab import files

# Step 2: Upload your Rosalind dataset
uploaded = files.upload()

# Step 3: Read the uploaded file
filename = list(uploaded.keys())[0]

with open(filename, 'r') as file:
    lines = file.read().strip().splitlines()

# Rosalind usually gives input in this format:
# Line 1: Pattern
# Line 2: Text
# Line 3: d (integer)
Pattern = lines[0].strip()
Text = lines[1].strip()
d = int(lines[2].strip())


# --- Helper Function ---
def hamming_distance(s1, s2):
    """
    Returns the number of positions where s1 and s2 differ.
    Assumes s1 and s2 are the same length.
    """
    mismatches = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            mismatches += 1
    return mismatches


# --- Main Function ---
def approximate_pattern_matching(Pattern, Text, d):
    """
    Finds all starting positions where Pattern appears in Text
    with at most d mismatches.
    """
    positions = []

    for i in range(len(Text) - len(Pattern) + 1):
        substring = Text[i:i+len(Pattern)]
        if hamming_distance(substring, Pattern) <= d:
            positions.append(i)

    return positions


# Step 4: Run the function on your dataset
result = approximate_pattern_matching(Pattern, Text, d)

# Step 5: Print results in one line, space-separated (Rosalind style)
print(" ".join(map(str, result)))
```

```python
#Rosalind 17

# Step 1: Import file upload tool
from google.colab import files

# Step 2: Upload your Rosalind dataset
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# Step 3: Read the uploaded file
with open(filename, 'r') as file:
    lines = file.read().strip().splitlines()

# Rosalind input format:
# Line 1: Text
# Line 2: k d (space-separated)
Text = lines[0].strip()
k, d = map(int, lines[1].strip().split())  # Read both integers from the same line


# --- Hamming distance function ---
def hamming_distance(s1, s2):
    mismatches = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            mismatches += 1
    return mismatches


# --- Neighbors function (recursive) ---
def Neighbors(Pattern, d):
    if d == 0:
        return {Pattern}
    if len(Pattern) == 1:
        return {'A', 'C', 'G', 'T'}

    Neighborhood = set()
    Suffix = Pattern[1:]
    SuffixNeighbors = Neighbors(Suffix, d)

    for text in SuffixNeighbors:
        if hamming_distance(Suffix, text) < d:
            for nucleotide in ['A', 'C', 'G', 'T']:
                Neighborhood.add(nucleotide + text)
        else:
            Neighborhood.add(Pattern[0] + text)

    return Neighborhood


# --- Main function ---
def most_frequent_kmers_with_mismatches(Text, k, d):
    kmer_counts = {}

    for i in range(len(Text) - k + 1):
        pattern = Text[i:i+k]
        neighbors = Neighbors(pattern, d)
        for neighbor in neighbors:
            if neighbor in kmer_counts:
                kmer_counts[neighbor] += 1
            else:
                kmer_counts[neighbor] = 1

    max_count = max(kmer_counts.values())

    most_frequent = [kmer for kmer, count in kmer_counts.items() if count == max_count]

    return most_frequent


# Step 4: Run the function
result = most_frequent_kmers_with_mismatches(Text, k, d)

# Step 5: Print results in one line, space-separated
print(" ".join(sorted(result)))
```
