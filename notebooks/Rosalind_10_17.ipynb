```python
#Rosalind 11

# Import the files module from google.colab
from google.colab import files

# Step 1: Upload your Rosalind dataset
# This will open a file upload dialog in Colab
uploaded = files.upload()

# Step 2: Read the uploaded file
# uploaded is a dictionary: keys are filenames, values are file contents
# We'll get the first uploaded file
filename = list(uploaded.keys())[0]

# Read DNA string and clean it
with open(filename, 'r') as file:
    dna_text = "".join(file.read().split()).upper()  # remove whitespace/newlines and capitalize

# Set k
k = 14  # replace with your Rosalind problem's k value

# Function to find all most frequent k-mers in a DNA string
def most_frequent_kmers(text, k):
    """
    Finds all k-length substrings (k-mers) in the DNA string `text`
    that occur the most frequently.

    Parameters:
    text (str): DNA sequence (string of A, C, G, T)
    k (int): length of k-mer

    Returns:
    list: list of most frequent k-mers (order does not matter)
    """
    # Dictionary to store counts of each k-mer
    kmer_counts = {}

    # Loop over every possible starting position of a k-mer
    for i in range(len(text) - k + 1):
        current_kmer = text[i:i+k]  # Extract k-mer of length k
        if current_kmer in kmer_counts:
            kmer_counts[current_kmer] += 1  # Increment count if already exists
        else:
            kmer_counts[current_kmer] = 1   # Initialize count if new k-mer

    # Find the maximum frequency among all k-mers
    max_count = max(kmer_counts.values())

    # Collect all k-mers that have the maximum frequency
    most_frequent = []
    for kmer, count in kmer_counts.items():
        if count == max_count:
            most_frequent.append(kmer)

    return most_frequent
result = most_frequent_kmers(dna_text, k)
print("Most frequent", k, "-mers:", result)
```

```python
#Rosalind 12

# Import the files module for uploading files in Colab
from google.colab import files

# Step 1: Upload your Rosalind dataset
uploaded = files.upload()
filename = list(uploaded.keys())[0]  # Take the first uploaded file

# Step 2: Read the Genome string from the file
with open(filename, 'r') as file:
    # Remove any whitespace or newlines, and convert to uppercase
    Genome = "".join(file.read().split()).upper()

# Step 3: Set your parameters (replace with values from your Rosalind problem)
k = 9  # length of k-mer
L = 591 # window length
t = 18  # minimum number of occurrences in window

# Step 4: Define the function to find (L, t)-clumps
def find_clumps(Genome, k, L, t):
    """
    Finds all distinct k-mers forming (L, t)-clumps in the Genome.

    Parameters:
    Genome (str): DNA sequence
    k (int): length of k-mer
    L (int): window size
    t (int): minimum number of times k-mer must appear in a window

    Returns:
    set: set of k-mers forming (L, t)-clumps
    """
    clumps = set()  # To store k-mers that form clumps

    # Slide a window of length L along the Genome
    for i in range(len(Genome) - L + 1):
        window = Genome[i:i+L]  # Extract current window
        window_counts = {}       # Dictionary to count k-mers in this window

        # Count k-mers in the current window
        for j in range(L - k + 1):
            kmer = window[j:j+k]
            window_counts[kmer] = window_counts.get(kmer, 0) + 1

        # Add k-mers meeting the threshold t to the clumps set
        for kmer, count in window_counts.items():
            if count >= t:
                clumps.add(kmer)

    return clumps

# Step 5: Run the function and print the results
result = find_clumps(Genome, k, L, t)
print("Number of clump-forming k-mers:", len(result))
print("Clump-forming k-mers:", result)
```
